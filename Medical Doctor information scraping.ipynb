```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
infolink=[]
driver=webdriver.Chrome(executable_path="F:\chromedriver.exe")
driver.get("https://www.okadoc.com/en-sa/search/result")
print(driver.title)
print(driver.current_url)
ll=driver.find_element_by_xpath('//*[@id="search-wrapper-card"]/div[1]')
#doctorInfoList= driver.find_elements_by_class_name('col-lg-6')
#for el in doctorInfoList:
    #ddd=el.find_elements_by_tag_name('a')[1]
 #   infolink.append(ddd.get_property('href'))
```

```python
listofLinks=[]
condition=True
j=0
while condition:
    doctorInfoList= driver.find_elements_by_class_name('col-lg-6')
    for el in doctorInfoList:
        ddd=el.find_elements_by_tag_name('a')[1]
        listofLinks.append(ddd.get_property('href'))
    try:
        #driver.find_element_by_xpath('//*[@id="search-wrapper-card"]/div[2]/ul/li[11]/a').click()
        driver.find_element_by_xpath('//*[@id="search-wrapper-card"]/div[2]/ul/li[12]').click()
    except:
        try:
            if(j<1):
                driver.find_element_by_xpath('//*[@id="search-wrapper-card"]/div[2]/ul/li[11]/a').click()
                j=+1
        except:
            condition= False
```

```python
from tqdm import tqdm
alldetails=[]
for i in tqdm(listofLinks):
    driver.get(i)
    try:
        Name=driver.find_element_by_xpath('//*[@id="js-main-profile-info"]/div/div[1]/div/div[1]/h1/span').text
    except:
        break
    Speciality=driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div/div[1]/div/div/nav/ol/li[3]/a').text
    Designation=driver.find_element_by_xpath('//*[@id="js-main-profile-info"]/div/div[1]/div/h6').text
    #driver.find_element_by_xpath()
    Image_URL=driver.find_element_by_xpath('//*[@id="js-main-profile"]/div[3]/div/img')
    image=Image_URL.get_attribute('src')
    try:
        AreasofExpertise=driver.find_element_by_xpath('//*[@id="symptoms-section"]/div[2]/div').text
    except:
        AreasofExpertise=False
    Clinic=driver.find_element_by_xpath('//*[@id="js-select-clinic-nav"]/div/div/div/div/div/div/div[2]/div/div[2]/ul/li/div/a').text
    Address=driver.find_element_by_xpath('//*[@id="clinic-loc-section"]/div[2]/div/div[2]/div[1]/div[2]/p').text
    try:
        Phone=driver.find_element_by_xpath('//*[@id="clinic-loc-section"]/div[2]/div/div[2]/div[3]/div[2]/p').text
    except:
        Phone=False
    try:
        Payments=driver.find_element_by_xpath('//*[@id="payments-section"]/div[2]/div').text
    except:
        Payments=False
    try:
        Language=driver.find_element_by_xpath('//*[@id="language-section"]/div[2]/div').text
    except:
        Language=False
        
    try:
        Education=driver.find_element_by_xpath('//*[@id="education-section"]/div[2]/div[1]').text
    except:
        Education=False
    try:
        RegistrationLicense=driver.find_element_by_xpath('//*[@id="registrations-section"]/div[2]/div/div/div/div[2]').text
    except:
        RegistrationLicense=False
    try:   
        dll=driver.find_elements_by_class_name('profile-detail-instance')
        l=0
        k=0

        dmm=[]
        for s in dll:
            try:
        
                d=dll[l].find_element_by_class_name('mb-3')
    
                kd=dll[l].find_elements_by_class_name('py-1')
                if d.text=="Bupa International":
                
                    for ss in kd:
                        dmm.append(ss.text)
                        k=k+1
                    else:
                        break
                l=l+1
            except:
                break

    except:
        dmm=False
    
    
    tempJ={'Name':Name,'Speciality':Speciality,'Designation':Designation,'Image_URL':image,
       'AreasofExpertise':AreasofExpertise,'Clinic':Clinic,'Address':Address,'Phone':Phone,
       'Payments':Payments,'Language':Language,'Education':Education,'RegistrationLicense':RegistrationLicense,'URL':i,'Bupa International':dmm}
    alldetails.append(tempJ)
```
